```python
import matplotlib.pyplot as plt
import pandas as pd
empdata={'empid':[1001,1002,1003,1004,1005],'empname':['amitava','nighat','dipankar','alpana','Nilanjan'],'salary':[10000,20000,30000,25000,50000]}
```

```python
empdata
```

```python
df=pd.DataFrame(empdata)
```

```python
df
```

```python
x=df['empid']
y=df['salary']
plt.bar(x,y,label='Anudip data',color='red')
plt.xlabel('EMPLOYEE ID')
plt.ylabel('SALARY of Emp')
plt.title('***********ANUDIP FOUNDATION************')
plt.legend()
plt.show()
```
